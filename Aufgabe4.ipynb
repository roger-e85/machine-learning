<h3>1. Welchen Trainingsalgorithmus für die lineare Regression können Sie verwenden, wenn Sie einen Trainingsdatensatz mit Millionen Merkmalen haben?</h3>

<p><strong>Antwort:</strong></p>

<p>
Für einen Trainingsdatensatz mit Millionen Merkmalen eignet sich besonders das
<strong>stochastische Gradientenverfahren (SGD)</strong> oder das
<strong>Mini-Batch-Gradientenverfahren</strong>.
</p>

<p><strong>Begründung:</strong></p>
<ul>
  <li>Die Normalengleichung ist bei sehr vielen Merkmalen zu rechenintensiv.</li>
  <li>Auch das Batch-Gradientenverfahren kann bei sehr grossen Datensätzen langsam werden.</li>
  <li>SGD und Mini-Batch Gradient Descent skalieren deutlich besser und sind für grosse Datensätze geeignet.</li>
</ul>

<p><strong>Kurzfassung:</strong></p>
<p>
Das stochastische Gradientenverfahren ist besonders geeignet, da es auch bei sehr vielen Merkmalen effizient trainieren kann.
</p>

<h3>2. Nehmen Sie an, dass die Merkmale in Ihrem Trainingsdatensatz unterschiedlich skaliert sind. Welche Algorithmen würden dadurch in Mitleidenschaft gezogen und in welcher Weise? Was können Sie dagegen tun?</h3>

<p><strong>Antwort:</strong></p>

<p>
Unterschiedlich skalierte Merkmale wirken sich vor allem negativ auf
<strong>gradientenbasierte Algorithmen</strong> aus.
</p>

<p><strong>Betroffene Algorithmen:</strong></p>
<ul>
  <li>Batch Gradient Descent</li>
  <li>Stochastic Gradient Descent (SGD)</li>
  <li>Mini-Batch Gradient Descent</li>
</ul>

<p><strong>Auswirkungen:</strong></p>
<ul>
  <li>Langsame Konvergenz</li>
  <li>Instabile Optimierung</li>
  <li>Zick-Zack-Bewegung im Gradientenverfahren</li>
</ul>

<p>
Wenn einige Merkmale sehr grosse Werte haben und andere sehr kleine,
dominiert das Merkmal mit den grossen Werten den Gradientenabstieg.
Dadurch dauert es länger, bis das Modell konvergiert.
</p>

<p><strong>Lösung:</strong></p>

<p>
Das Problem kann durch <strong>Feature-Scaling</strong> behoben werden:
</p>

<ul>
  <li>Standardisierung (StandardScaler)</li>
  <li>Min-Max-Skalierung (MinMaxScaler)</li>
</ul>

<p><strong>Kurzfassung:</strong></p>
<p>
Gradientenbasierte Algorithmen werden durch unterschiedlich skalierte Merkmale
negativ beeinflusst. Durch Feature-Scaling können die Merkmale auf eine
vergleichbare Skala gebracht werden, wodurch das Training stabiler und schneller wird.
</p>

<h3>3. Kann das Gradientenverfahren bei einem logistischen Regressionsmodell in einem lokalen Minimum stecken bleiben?</h3>

<p><strong>Antwort:</strong></p>

<p>
Nein, das Gradientenverfahren kann bei der logistischen Regression nicht in einem lokalen Minimum stecken bleiben.
</p>

<p><strong>Begründung:</strong></p>

<p>
Die Kostenfunktion der logistischen Regression ist <strong>konvex</strong>. 
Das bedeutet, dass es nur ein <strong>globales Minimum</strong> gibt und keine lokalen Minima existieren.
</p>

<p>
Dadurch konvergiert das Gradientenverfahren immer zur optimalen Lösung, 
vorausgesetzt, die Lernrate ist sinnvoll gewählt und der Algorithmus läuft lange genug.
</p>

<p><strong>Kurzfassung:</strong></p>

<p>
Da die Kostenfunktion der logistischen Regression konvex ist, kann das Gradientenverfahren 
nicht in lokalen Minima stecken bleiben und findet immer das globale Minimum.
</p>

<h3>4. Führen alle Algorithmen für das Gradientenverfahren zum gleichen Modell, vorausgesetzt, sie laufen lange genug?</h3>

<p><strong>Antwort:</strong></p>

<p>
Ja, grundsätzlich führen alle Varianten des Gradientenverfahrens zum gleichen Modell, 
sofern sie lange genug laufen und eine geeignete Lernrate verwendet wird.
</p>

<p><strong>Begründung:</strong></p>

<p>
Die verschiedenen Gradientenverfahren unterscheiden sich hauptsächlich in der Art, 
wie sie die Gradienten berechnen:
</p>

<ul>
  <li><strong>Batch Gradient Descent</strong> verwendet den gesamten Trainingsdatensatz</li>
  <li><strong>Stochastic Gradient Descent (SGD)</strong> verwendet jeweils ein einzelnes Trainingsbeispiel</li>
  <li><strong>Mini-Batch Gradient Descent</strong> verwendet kleine Teilmengen der Trainingsdaten</li>
</ul>

<p>
Obwohl diese Verfahren unterschiedlich arbeiten, optimieren sie alle dieselbe 
Kostenfunktion. Daher konvergieren sie bei genügend langer Laufzeit zur gleichen 
optimalen Lösung.
</p>

<p>
Unterschiede bestehen hauptsächlich in:
</p>

<ul>
  <li>Geschwindigkeit der Konvergenz</li>
  <li>Stabilität der Optimierung</li>
  <li>Speicherbedarf</li>
</ul>

<p><strong>Kurzfassung:</strong></p>

<p>
Alle Gradientenverfahren führen zur gleichen optimalen Lösung, sofern sie lange 
genug laufen und eine geeignete Lernrate gewählt wird. Sie unterscheiden sich 
hauptsächlich in Geschwindigkeit und Stabilität.
</p>

<h3>5. Nehmen Sie an, Sie verwenden das Batch-Gradientenverfahren und plotten den Validierungsfehler in jeder Epoche. Was passiert vermutlich, wenn der Validierungsfehler ständig steigt? Wie können Sie dies beheben?</h3>

<p><strong>Antwort:</strong></p>

<p>
Wenn der Validierungsfehler ständig steigt, während der Trainingsfehler weiter sinkt,
liegt vermutlich <strong>Overfitting</strong> vor.
</p>

<p><strong>Erklärung:</strong></p>

<p>
Das Modell passt sich immer stärker an die Trainingsdaten an und lernt dabei auch 
Zufälligkeiten und Rauschen in den Daten. Dadurch verschlechtert sich die Fähigkeit 
des Modells, auf neuen, unbekannten Daten zu generalisieren. Dies zeigt sich durch 
einen steigenden Validierungsfehler.
</p>

<p><strong>Mögliche Lösungen:</strong></p>

<ul>
  <li>Early Stopping (Training frühzeitig beenden)</li>
  <li>Regularisierung einsetzen (z.B. Ridge oder Lasso)</li>
  <li>Modell vereinfachen (weniger Features oder geringerer Polynomialgrad)</li>
  <li>Mehr Trainingsdaten verwenden</li>
</ul>

<p><strong>Kurzfassung:</strong></p>

<p>
Ein ständig steigender Validierungsfehler deutet auf Overfitting hin. Dieses Problem 
kann durch Early Stopping, Regularisierung, ein einfacheres Modell oder mehr 
Trainingsdaten behoben werden.
</p>

<h3>6. Ist es eine gute Idee, das Mini-Batch-Gradientenverfahren sofort zu unterbrechen, sobald der Validierungsfehler steigt?</h3>

<p><strong>Antwort:</strong></p>

<p>
Nein, es ist in der Regel keine gute Idee, das Mini-Batch-Gradientenverfahren sofort zu 
unterbrechen, sobald der Validierungsfehler steigt.
</p>

<p><strong>Begründung:</strong></p>

<p>
Beim Mini-Batch-Gradientenverfahren werden nur kleine Teilmengen der Trainingsdaten 
verwendet. Dadurch schwankt der Validierungsfehler stärker als beim Batch-Gradientenverfahren. 
Ein kurzfristiger Anstieg des Validierungsfehlers ist daher normal und bedeutet nicht 
zwangsläufig, dass das Modell overfittet.
</p>

<p>
Wenn das Training sofort abgebrochen wird, besteht die Gefahr, dass das Modell zu früh 
stoppt und noch nicht optimal trainiert ist.
</p>

<p><strong>Bessere Vorgehensweise:</strong></p>

<ul>
  <li>Early Stopping mit Geduld (Patience) verwenden</li>
  <li>Training erst abbrechen, wenn der Validierungsfehler über mehrere Epochen hinweg steigt</li>
</ul>

<p><strong>Kurzfassung:</strong></p>

<p>
Nein, da der Validierungsfehler beim Mini-Batch-Gradientenverfahren schwanken kann. 
Stattdessen sollte Early Stopping mit Geduld verwendet werden, um ein vorzeitiges 
Beenden des Trainings zu vermeiden.
</p>

<h3>7. Welcher der besprochenen Algorithmen für das Gradientenverfahren erreicht die Umgebung der optimalen Lösung am schnellsten? Welcher konvergiert? Wie können Sie auch die übrigen konvergieren lassen?</h3>

<p><strong>Antwort:</strong></p>

<p>
Das <strong>Stochastic Gradient Descent (SGD)</strong> erreicht die Umgebung der optimalen 
Lösung am schnellsten. Dies liegt daran, dass bei jedem Schritt nur ein einzelnes 
Trainingsbeispiel verwendet wird, wodurch sehr schnelle Updates möglich sind.
</p>

<p>
Das <strong>Batch Gradient Descent</strong> konvergiert hingegen am zuverlässigsten zum 
exakten Minimum, da bei jedem Schritt der gesamte Trainingsdatensatz verwendet wird. 
Dies führt zu stabileren, aber langsameren Updates.
</p>

<p>
Das <strong>Mini-Batch Gradient Descent</strong> stellt einen Kompromiss zwischen beiden 
Verfahren dar. Es ist schneller als Batch Gradient Descent und stabiler als SGD.
</p>

<p><strong>Wie können die übrigen Verfahren konvergieren?</strong></p>

<ul>
  <li>Lernrate im Verlauf des Trainings reduzieren</li>
  <li>Lernraten-Scheduling verwenden</li>
  <li>Training über genügend viele Iterationen laufen lassen</li>
</ul>

<p><strong>Kurzfassung:</strong></p>

<p>
SGD erreicht die Umgebung der optimalen Lösung am schnellsten. Batch Gradient Descent 
konvergiert am zuverlässigsten. Durch Reduzieren der Lernrate können auch SGD und 
Mini-Batch Gradient Descent zur optimalen Lösung konvergieren.
</p>

<h3>10. Welche Gründe sprechen für folgende Verfahren?</h3>

<p><strong>Ridge-Regression anstatt einer einfachen linearen Regression?</strong></p>

<p>
Die Ridge-Regression verwendet eine Regularisierung, um Overfitting zu vermeiden.
Dabei werden die Modellgewichte verkleinert, wodurch das Modell stabiler wird und
besser auf neue Daten generalisiert.
</p>

<ul>
  <li>Reduziert Overfitting</li>
  <li>Verbessert Generalisierung</li>
  <li>Stabilisiert das Modell bei korrelierten Merkmalen</li>
</ul>

<p><strong>Lasso anstelle einer Ridge-Regression?</strong></p>

<p>
Die Lasso-Regression kann zusätzlich zur Regularisierung auch eine
<strong>Feature-Selektion</strong> durchführen. Dabei werden unwichtige Merkmale
automatisch entfernt, indem deren Gewichte auf Null gesetzt werden.
</p>

<ul>
  <li>Automatische Feature-Selektion</li>
  <li>Einfacheres Modell</li>
  <li>Bessere Interpretierbarkeit</li>
</ul>

<p><strong>Elastic Net anstelle von Lasso?</strong></p>

<p>
Elastic Net kombiniert Ridge- und Lasso-Regression. Dies ist besonders hilfreich,
wenn viele Merkmale stark miteinander korreliert sind. Elastic Net ist stabiler
als Lasso und behält gleichzeitig die Vorteile der Feature-Selektion.
</p>

<ul>
  <li>Kombination aus Ridge und Lasso</li>
  <li>Stabiler bei korrelierten Merkmalen</li>
  <li>Ermöglicht weiterhin Feature-Selektion</li>
</ul>

<p><strong>Kurzfassung:</strong></p>

<p>
Ridge reduziert Overfitting, Lasso ermöglicht zusätzlich Feature-Selektion, und
Elastic Net kombiniert die Vorteile beider Verfahren und ist besonders stabil bei
korrelierten Merkmalen.
</p>

<h3>11. Angenommen, Sie möchten Bilder als innen/aussen und Tag/Nacht klassifizieren. Sollten Sie zwei Klassifikatoren mit logistischer Regression oder einen Klassifikator mit Softmax-Regression erstellen?</h3>

Die Aufgabe beinhaltet vier mögliche Klassen:

<p><strong>Kurzfassung:</strong></p>

<p>

Softmax-Regression-Klassifikator die geeignetere Wahl.
</p>

<h2>Aufgabe 12</h2>
<p>Implementieren Sie das Batch-Gradientenverfahren mit Early Stopping für
die Softmax-Regression (ohne Scikit-Learn).</p>


<h3>Daten laden und vorbereiten</h3>

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X = iris["data"]
y = iris["target"]

# Optional: nur 2 Merkmale verwenden, damit es einfacher bleibt
X = X[:, [2, 3]]  # petal length, petal width

# Bias-Term hinzufügen
X_bias = np.c_[np.ones((len(X), 1)), X]

# Train / Validation / Test
X_train_val, X_test, y_train_val, y_test = train_test_split(
    X_bias, y, test_size=0.2, random_state=42, stratify=y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val, test_size=0.2, random_state=42, stratify=y_train_val
)

n_inputs = X_train.shape[1]
n_classes = len(np.unique(y))

# One-Hot-Encoding für die Zielwerte
def to_one_hot(y, n_classes):
    one_hot = np.zeros((len(y), n_classes))
    one_hot[np.arange(len(y)), y] = 1
    return one_hot

Y_train_one_hot = to_one_hot(y_train, n_classes)
Y_val_one_hot = to_one_hot(y_val, n_classes)
Y_test_one_hot = to_one_hot(y_test, n_classes)
```

<h3>Softmax-Funktion</h3>

```python
def softmax(logits):
    # numerisch stabil
    exp_shifted = np.exp(logits - np.max(logits, axis=1, keepdims=True))
    return exp_shifted / np.sum(exp_shifted, axis=1, keepdims=True)
```

<h3>Kreuzentropie-Verlust</h3>

```python
def cross_entropy_loss(y_true_one_hot, y_pred_proba, epsilon=1e-15):
    y_pred_proba = np.clip(y_pred_proba, epsilon, 1 - epsilon)
    return -np.mean(np.sum(y_true_one_hot * np.log(y_pred_proba), axis=1))
```

<h3>Vorhersagefunktion</h3>

```python
def predict(X, theta):
    logits = X @ theta
    proba = softmax(logits)
    return np.argmax(proba, axis=1)
```

<h3>Training mit Batch Gradient Descent + Early Stopping</h3>

```python
np.random.seed(42)
theta = np.random.randn(n_inputs, n_classes) * 0.01

eta = 0.1                  # Lernrate
n_epochs = 5000
patience = 50              # wie viele Epochen ohne Verbesserung erlaubt sind
best_val_loss = np.inf
best_theta = None
epochs_without_improvement = 0

train_losses = []
val_losses = []

for epoch in range(n_epochs):
    # Vorwärtsdurchlauf Training
    logits_train = X_train @ theta
    y_proba_train = softmax(logits_train)

    # Verlust Training
    train_loss = cross_entropy_loss(Y_train_one_hot, y_proba_train)
    train_losses.append(train_loss)

    # Gradient berechnen (Batch Gradient Descent)
    error = y_proba_train - Y_train_one_hot
    gradients = (1 / len(X_train)) * X_train.T @ error

    # Parameter-Update
    theta -= eta * gradients

    # Validierung
    logits_val = X_val @ theta
    y_proba_val = softmax(logits_val)
    val_loss = cross_entropy_loss(Y_val_one_hot, y_proba_val)
    val_losses.append(val_loss)

    # Early Stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_theta = theta.copy()
        epochs_without_improvement = 0
    else:
        epochs_without_improvement += 1

    if epoch % 100 == 0:
        print(f"Epoche {epoch:4d} | Trainingsverlust: {train_loss:.4f} | Validierungsverlust: {val_loss:.4f}")

    if epochs_without_improvement >= patience:
        print(f"\nEarly Stopping bei Epoche {epoch}")
        break

# Beste Parameter wiederherstellen
theta = best_theta
```

<h3>Auswertung</h3>

```python
y_train_pred = predict(X_train, theta)
y_val_pred = predict(X_val, theta)
y_test_pred = predict(X_test, theta)

train_acc = np.mean(y_train_pred == y_train)
val_acc = np.mean(y_val_pred == y_val)
test_acc = np.mean(y_test_pred == y_test)

print("\nEndergebnisse:")
print(f"Trainingsgenauigkeit:   {train_acc:.4f}")
print(f"Validierungsgenauigkeit:{val_acc:.4f}")
print(f"Testgenauigkeit:        {test_acc:.4f}")
```

<p>Für die Softmax-Regression wurde das Batch-Gradientenverfahren ohne Scikit-Learn mit numpy implementiert. Dabei wurden die Trainingsdaten in Trainings-, Validierungs- und Testdaten unterteilt. In jeder Epoche wurde der Gradient der Kreuzentropie-Kostenfunktion über den gesamten Trainingsdatensatz berechnet und die Parameter aktualisiert. Zur Vermeidung von Overfitting wurde Early Stopping eingesetzt: Sobald sich der Validierungsverlust über mehrere Epochen hinweg nicht mehr verbesserte, wurde das Training beendet und die besten bisher gefundenen Parameter wurden übernommen.</p>
